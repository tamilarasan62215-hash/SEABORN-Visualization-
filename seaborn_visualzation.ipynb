# univarite analysis

kde=(kernul density estimation)
hue=(give a differenciate from dataset)
diag_kind=(diagnal line)kde(or)hue

```python
import seaborn as sns 
sns.histplot(dataset["ssc_p"],kde=True)# histogram (frequency or reputation)for unick colum
plt.show()
```

```python
sns.histplot(dataset["ssc_p"],kde=True,stat="density",line_kws={"color":"red"})
```

# Bivariate analysis (Distribution plot)

```python
#joinplot (combaine the two column dataset values in single graph)
sns.jointplot(x="ssc_p",y="salary",data=dataset)
plt.show()                       #(scattered [points] and histogram)
```

# hex(hexagan)plot

```python
sns.jointplot(x="ssc_p",y="salary",data=dataset,kind="hex")
plt.show()
# finde the high reputation of color variance like (heat method) dark color high reputation
# 10th stuents 51 or 52 mark grtting persons are grtting salary nearly 3Lack so may studens 
```

# kde(kernal density estimation)

```python
sns.jointplot(x="ssc_p",y="salary",data=dataset,kind="kde")
plt.show()    # data present between 40 to 90 mark getting salary 1.5 l to 5l 
```

# pairplot 

```python
sns.pairplot(dataset,hue="gender",diag_kind="kde",kind="scatter",palette="husl")
plt.show()
# hue (differencate)
#diag_kind (graph view (kde) or (hist) or (hex) types)
#kind(values spread point)
#palette=("husl") # color different
```

```python
# combaine data view two dataset values 
```

# Categorical plot (qualitative datas)

```python
sns.stripplot(x="ssc_b",y="ssc_p",data=dataset)
plt.show()
```

```python
sns.stripplot(x="gender",y="ssc_p",data=dataset,jitter=True)# jitter=true(avoid overlap points maximum jointng litle split the point )
plt.show()
```

```python
sns.stripplot(x="gender",y="ssc_p",data=dataset) # points in small overlap  points are joined
plt.show()
```

```python
sns.swarmplot(x="gender",y="ssc_p",data=dataset)
plt.show # points are not overlap point -not points are join
```

# Distribution observation    (Violinplot)

```python
sns.violinplot(x="gender",y="ssc_p",data=dataset) # box in black refers Q1,Q2,Q3,Q4 data present
plt.show()
```

```python
sns.violinplot(x="ssc_b",y="ssc_p",data=dataset,hue="gender")
plt.show()
```

```python
sns.violinplot(x="specialisation",y="salary",data=dataset,hue="gender")
plt.show()
```

```python
sns.violinplot(x="specialisation",y="salary",data=dataset)# with out gender
plt.show()
```
